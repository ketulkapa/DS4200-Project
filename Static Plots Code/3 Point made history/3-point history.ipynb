```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data
df = pd.read_csv('Playoffs.csv')

# Calculate average 3-pointers made and attempted (per game by season)
three_point_stats = df.groupby('year').agg({
    'GP': 'sum',
    'FG3M': 'sum',
    'FG3A': 'sum'
}).reset_index()

# Calculate per game averages
three_point_stats['FG3M_per_game'] = three_point_stats['FG3M'] / three_point_stats['GP']
three_point_stats['FG3A_per_game'] = three_point_stats['FG3A'] / three_point_stats['GP']

# Plot both made and attempted 3-pointers
plt.figure(figsize = (9, 7))

plt.plot(three_point_stats['year'], three_point_stats['FG3M_per_game'], 
         marker = 'o', label = '3-Pointers Made per Game')

plt.plot(three_point_stats['year'], three_point_stats['FG3A_per_game'], 
         marker = 's', label = '3-Pointers Attempted per Game')

# Format the visualization
plt.title('NBA Playoffs: 3-Point Shooting Evolution (2012-2024)')
plt.xlabel('Season')
plt.ylabel('3-Pointers per Game')
plt.grid(True)
plt.legend()
plt.xticks(rotation = 45)
plt.tight_layout()

# Add text annotation explaining the trend
avg_2012 = round(three_point_stats.iloc[0]['FG3M_per_game'], 2)
avg_2023 = round(three_point_stats.iloc[-1]['FG3M_per_game'], 2)
pct_increase = round(((avg_2023 - avg_2012) / avg_2012) * 100)

annotation_text = f"3 Point Made per game increased by {pct_increase}%\nfrom {avg_2012} (2012-13) to {avg_2023} (2023-24)"

plt.text(0.01, 1, annotation_text, transform = plt.gca().transAxes, 
         verticalalignment = 'top', bbox = dict(facecolor = 'white', alpha = 0.8))

# Save the plot
plt.savefig('nba_playoff_3pt_trend.png')

# Show the plot
plt.show()
```

### Finding
Some key observation our team found is there is a significant increase for 3 point attempt per game in play-off from 2.0 in 12-13 season to 3.2 for 23-24 season. 3 point made per game is also increase by 68% within the past 12 seasons. The rise of 3-point statistics showing a shift in gameplay strategies over time. Player are adapting 3-point shooting as a primary score method instead of what they did in the past (near basket and pull up jumpers). This has also change the team dynamic of the transformation of defensive strategies, spacing on the court, and player roles, emphasizing shooters and perimeter play. Due to this transformation, the 3 point percentage is likely to be slightly higher but the trend may be remaining a stable increase because there are all kind of different defense approch has been change to adapt the transofrmation of scoring technique.
